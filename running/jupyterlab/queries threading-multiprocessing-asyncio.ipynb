In next run, command option `-i` tells the program that input files are inside path `../queries/` and every file have extension `.in`. The option `-o` tells the program that output files must be situated in the same input directory but with extension `.out`. The `-e` option tells the engine to use the provided token (with option `**user`), a waiting factor of 7 seconds and intructs to use a different cache for each query with option `**reset-cache`. To avoid overload the server the sub-queries are emitted after waiting `random.triangular(delay, delay * waiting_factor, delay)` seconds where delay depends on if the user is authenticated or not an in this case is 2 seconds. The function `random.triangular`[^1] in this case gives a random number between `delay` and `delay * waiting_factor` in a linear distribution where `delay` have the maximum probability. With option `--log-files` it is given a log file path `../debug-reset-cache.log` with a verbose level of `debug`. In this case the cache is not persistent. The queries inside each file are given in the output. To re-run the notebook change the token given to  **user option for one of your own.
[^1]: <https://docs.python.org/3.7/library/random.html\#random.triangular>
